```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.pardir))) 
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.pardir)))
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (adjust path as needed)
df = pd.read_csv('../data/raw.csv', parse_dates=['date'])

# Ensure 'date' column is datetime (usually handled by parse_dates but just in case)
df['date'] = pd.to_datetime(df['date'], errors='coerce')


# 🗓️ Daily publication frequency
df['day'] = df['date'].dt.date
df['day'].value_counts().sort_index().plot(figsize=(12, 4), title='Daily Article Frequency')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# 🕒 Hourly publication pattern
df['hour'] = df['date'].dt.hour
df['hour'].value_counts().sort_index().plot(kind='bar', figsize=(10, 3), title='Articles by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
